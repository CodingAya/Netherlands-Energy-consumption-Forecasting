```python
directory = 'csv_folder'
file_extension = '.csv'

for filename in os.listdir(directory):
    if filename.endswith(file_extension):
        file_path = os.path.join(directory, filename)
        
        # Extract year from filename (you might need to adjust this based on your file naming pattern)
        year = int(filename.split('_')[-1].split('.')[0])  # Assuming the year is part of the filename separated by underscores
        
        # Read the CSV file into a DataFrame
        df = pd.read_csv(file_path)
        
        # Add 'year' column to the DataFrame
        df['year'] = year
        
        # Save the modified DataFrame back to the CSV file
        df.to_csv(file_path, index=False)
```

```python
import matplotlib.gridspec as gridspec

# Assuming df_e_final is your DataFrame

specs = {'markersize': 8, 'markerfacecolor': 'w', 'linewidth': 2}

f = plt.figure()

gs = gridspec.GridSpec(3, 2)  # Adjust the grid specification as needed
f.set_figwidth(13)
f.set_figheight(10)
plt.suptitle('Electricity Consumption in Netherlands', fontsize=20)

# Plot 1: Total yearly energy consumption
ax1 = f.add_subplot(gs[0, 0])
ax1.plot(df_e_final.groupby('year')['annual_consume'].sum().index, df_e_final.groupby('year')['annual_consume'].sum().values, '-o', **specs)
plt.ylabel('Energy consumption (kWh)')
plt.title('Total yearly energy consumption')

# Plot 2: Total number of connections
ax2 = f.add_subplot(gs[0, 1])
ax2.plot(df_e_final.groupby('year')['num_connections'].sum().index, df_e_final.groupby('year')['num_connections'].sum().values, '-og', **specs)
plt.ylabel('Count')
plt.title('Total number of connections')

# Plot 3: Total yearly consumption during the low tarif hours
ax3 = f.add_subplot(gs[1, 0])
ax3.plot(df_e_final.groupby('year')['annual_consume_lowtarif_perc'].sum().index, df_e_final.groupby('year')['annual_consume_lowtarif_perc'].sum().values, '-o', **specs)
plt.ylabel('Energy consumption (kWh)')
plt.title('Total yearly energy consumption during the low tarif hours')

# Plot 4: Total number of perc_of_active_connections
ax4 = f.add_subplot(gs[1, 1])
ax4.plot(df_e_final.groupby('year')['perc_of_active_connections'].sum().index, df_e_final.groupby('year')['perc_of_active_connections'].sum().values, '-og', **specs)
plt.ylabel('Count')
plt.title('Total number of active connections')

# Plot 5: Total number of delivery_perc
ax5 = f.add_subplot(gs[2, 0])
ax5.plot(df_e_final.groupby('year')['delivery_perc'].sum().index, df_e_final.groupby('year')['delivery_perc'].sum().values, '-om', **specs)
plt.ylabel('Net consumption of electricity (kWh)')
plt.xlabel('Year')
plt.title('Total yearly net consumption of electricity')

# Plot 6: Total number of smartmeters
ax6 = f.add_subplot(gs[2, 1])
ax6.plot(df_e_final.groupby('year')['smartmeter_perc'].sum().index, df_e_final.groupby('year')['smartmeter_perc'].sum().values, '-oc', **specs)
plt.ylabel('Count')
plt.xlabel('Year')
plt.title('Total number of smartmeters')

gs.update(wspace=0.4, hspace=0.6)  # Adjust the spacing between subplots
plt.tight_layout()
plt.show()
```
